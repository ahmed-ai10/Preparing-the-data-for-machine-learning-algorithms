```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
```

```python
housing=pd.read_csv('housing.csv')
housing
```

```python
coor_matrix = housing.corr(numeric_only=True)
```

```python
coor_matrix['median_house_value'].sort_values(ascending=False)
```

```python
housing['rooms_per_house'] = housing['total_rooms']/housing['households']
```

```python
def roun(x):
    return round(x)
housing['rooms_per_house']=housing['rooms_per_house'].apply(roun)
```

```python
housing.isna().sum()
```

```python
imputer =SimpleImputer(strategy="median")
housing_num = housing.select_dtypes(include=[np.number])
```

```python
imputer.fit(housing_num)
```

```python
housing_num=pd.DataFrame(
    imputer.fit_transform(housing_num),
    columns=housing_num.columns)
```

```python
housing[housing_num.columns] = housing_num
```

```python
housing_cat = housing[['ocean_proximity']]
cat_encoder= OneHotEncoder()
housing_cat_enc= cat_encoder.fit_transform(housing_cat)
```

```python
housing_cat=pd.DataFrame(
    housing_cat_enc.toarray(),
    columns= cat_encoder.get_feature_names_out())
```

```python
housing = pd.concat([housing,housing_cat],axis=1)
```

```python
housing=housing.drop('ocean_proximity',axis=1)
housing.columns = housing.columns.str.replace('ocean_proximity_','',regex=False)
```

```python
housing['bedroms_ratio'] = housing['total_bedrooms']/housing['total_rooms']
housing['people_per_house'] = housing['population']/housing['households']
```

```python
def roun(x):
    return round(x)
housing['people_per_house']=housing['people_per_house'].apply(roun)
```

```python
coor_matrix = housing.corr(numeric_only=True)
coor_matrix['median_house_value'].sort_values(ascending=False)
```
